```python
import pandas as pd
import folium
```

```python
temples_df = pd.read_csv('ChurchofJesusChristTemples.csv', encoding = "ISO-8859-1")
temples_df.sample(10)
```

```python
temples_df = temples_df[['Temple', 'Status', 'Latitude', 'Longitude']]
temples_df.sample(10)
```

```python
# Creating the map and adding points to it
m = folium.Map(location=[temples_df.Latitude.mean(), temples_df.Longitude.mean()], zoom_start=5, control_scale=True)
colordict = {
   'Dedicated' : 'green',  
   'Under Construction' : 'purple', 
   'Announced' : 'orange',
}

for index, location_info in temples_df.iterrows():
    value = location_info['Status']
    color = colordict[value]  
    folium.Marker([location_info["Latitude"], location_info["Longitude"]], 
                    icon=folium.Icon(color=color), popup=location_info["Temple"]).add_to(m)
```

```python
m.save('map.html')
```
