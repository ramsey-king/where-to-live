```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import os
import openrouteservice
```

```python
temples_df = pd.read_csv('ChurchofJesusChristTemples.csv', encoding = "ISO-8859-1")
temples_df.sample(10)
```

```python
temples_df = temples_df[['Temple', 'Status', 'Latitude', 'Longitude']]
temples_df.sample(10)
```

```python
# Creating the map and adding points to it
m = folium.Map(location=[temples_df.Latitude.mean(), temples_df.Longitude.mean()], zoom_start=5, control_scale=True)

# tCluster = MarkerCluster(name='Temples').add_to(m)
colordict = {
   'Dedicated' : 'green',  
   'Under Construction' : 'purple', 
   'Announced' : 'orange',
}

for index, location_info in temples_df.iterrows():
    value = location_info['Status']
    color = colordict[value]  
    folium.Marker([location_info["Latitude"], location_info["Longitude"]], 
                    icon=folium.Icon(prefix='fa', icon='fort-awesome',color=color), tooltip=location_info["Temple"]+': '+value).add_to(m)

# folium.LayerControl().add_to(m)
```

```python
meetinghouses_df = pd.read_csv('lds_meetinghouses_20230308.csv')
meetinghouses_df.head()
```

```python
meetinghouses_df = meetinghouses_df[['name', 'country', 'latitude', 'longitude']]
meetinghouses_df.head()
```

```python
meetinghouses_df.shape
```

```python
meetinghouses_df.name.isnull().sum()
```

```python
meetinghouses_df = meetinghouses_df[~meetinghouses_df['name'].isnull()]
meetinghouses_df.shape
```

```python
mCluster = MarkerCluster(name='Meetinghouses').add_to(m)

for index, location_info in meetinghouses_df.iterrows():
    folium.Marker([location_info["latitude"], location_info["longitude"]], 
                    icon=folium.Icon(icon="university", prefix='fa'), tooltip=location_info["name"]).add_to(mCluster)

folium.LayerControl().add_to(m)

m.save('map.html')
```
